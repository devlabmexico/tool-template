```python
from os import environ
from os.path import join, splitext

input_folder = environ.get(
    'CROSSCOMPUTE_INPUT_FOLDER', 'tests/standard/input')
output_folder = environ.get(
    'CROSSCOMPUTE_OUTPUT_FOLDER', 'tests/standard/output')
```

```python
import pandas as pd

input_data_path = join(input_folder, 'input-data.csv')
output_data_path = join(output_folder, 'output-data.json')

input_df = pd.read_csv(input_data_path, index_col='id')

input_df.to_json(output_data_path, orient='index')

input_df
```
